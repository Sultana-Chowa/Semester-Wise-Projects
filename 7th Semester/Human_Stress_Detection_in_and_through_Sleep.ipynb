```python
X=df.iloc[:, 0:-1]
y= df.iloc[:, -1:]
y
```

```python
from mlxtend.preprocessing import minmax_scaling
X_scaled = minmax_scaling(X, columns=['snoring rate', 'respiration rate', 'body temperature',
                                                     'limb movement', 'blood oxygen', 'eye movement', 'sleeping hours',
                                                     'heart rate'])
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier

dtc = DecisionTreeClassifier()
lr = LogisticRegression()
rfc = RandomForestClassifier()
knn = KNeighborsClassifier()
```

```python
models = [dtc, lr,rfc,  knn]
acc_scores = []
for model in models:
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    acc_model = round(accuracy_score(y_pred, y_val) * 100, 2)
    acc_scores.append(acc_model)
acc_scores
```

```python
model_name = ['Decision Tree', 'Logistic Regression', 'Random Forest', 'KNN or k-Nearest Neighbors']
models_acc = pd.DataFrame({'Model name': model_name, 'Accuracy scores': acc_scores})
models_acc
```

```python
from sklearn.model_selection import cross_val_score

cv_mean_scores = []
for model in models:
    scores = cross_val_score(model, X, y,
                              cv=5,
                              scoring='accuracy')
    cv_mean_scores.append(scores.mean())
```

```python
models_val = pd.DataFrame({'Model name': model_name, 'Cross validation mean scores': cv_mean_scores})
models_val = models_val.sort_values(by='Cross validation mean scores', ascending=False)
models_val
```

```python
from sklearn.metrics import confusion_matrix  # For find confusion matrix
from sklearn.metrics import classification_report  # For print evaluation report
from sklearn.metrics import roc_curve  # For drawing ROC curve
from sklearn.metrics import auc  # For find AUC
# Prediction of test dataset:
y_pred_lr = lr.predict(X_val)

# Find confusion matrix for this model:
confusion_matrix_LR = confusion_matrix(y_val, y_pred_lr)

# Plot confusion matrix with Heatmap:
cm_dataframe_LR = pd.DataFrame(confusion_matrix_LR, index=['0', '1', '2', '3', '4'], columns=['0', '1', '2', '3', '4'])
sns.heatmap(cm_dataframe_LR, annot=True, annot_kws={"size": 18}, fmt="d")
plt.title("Logistic Regression")
plt.ylabel('Actual Classes')
plt.xlabel('Predicted Classes')
plt.show()
```

```python
report_LR = pd.DataFrame(classification_report(y_val, y_pred_lr,
                                               output_dict=True,
                                               target_names=['0', '1', '2', '3', '4']))
report_LR
```
